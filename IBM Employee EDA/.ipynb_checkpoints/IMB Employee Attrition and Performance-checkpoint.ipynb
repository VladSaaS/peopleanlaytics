# IBM Employee Attrition and Performance Analysis

The scope of this analysis is to determine relationships between compensation and performance.
We will begin by importing the necessary libraries and from there explore the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Lets now load the dataset into a Pandas DataFrame:

```python
df = pd.read_csv('IBM.csv')
```

Next up will be getting basic informatio nabout the data set, getting descriptive statistics and also checking for missing values.

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

It looks like the dataset is generally tidy with no missing values and we are able to review the EDA data for further examination. Next up, I am interested to find out the spread of each numerical feature and look for correlations through a heat map. We will use histograms as well as a correlation heatmap.

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64'])
numerical_cols.hist(bins=15, figsize=(20, 15))
plt.show()
```

```python
# Select numerical columns for correlation analysis (excluding EmployeeCount and StandardHours)
numerical_cols = df.select_dtypes(include=['int64', 'float64'])
numerical_cols = numerical_cols.drop(['EmployeeCount', 'StandardHours'], axis=1)

# Calculate the correlation matrix
correlation_matrix = numerical_cols.corr()

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap (Excluding EmployeeCount and StandardHours)')
plt.show()
```
