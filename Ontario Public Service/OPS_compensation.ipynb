Date Accessed: October 12, 2023

```python
# Importing py libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Processing dataset as a py dataframe
df = pd.read_csv('ops_workforce_demographic_dataset_-_2015_to_2023_english.csv', encoding='ISO-8859-1')

# dropping the first two rows so that column titles are at index 0
df.columns = df.iloc[1]
df = df.drop([0, 1]).reset_index(drop=True)
```

```python
# Exploratory data analysis (EDA)
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# Converting Average Salary ($), Median Salary ($), Female Average Annual Salary and Male Average Annual Salary features to float

df['Average Salary ($)'] = df['Average Salary ($)'].astype(float)
df['Median Salary ($)'] = df['Median Salary ($)'].astype(float)
df['Female Average Annual Salary'] = df['Female Average Annual Salary'].astype(float)
df['Male Average Annual Salary'] = df['Male Average Annual Salary'].astype(float)
```

```python
# Plotting compensation features transformed above
df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the index for time-based plotting.
df.set_index('Date', inplace=True)

# Plotting the trends over time
plt.figure(figsize=(12, 6))

# Plotting Average Salary ($) over time
plt.subplot(221)
plt.plot(df.index, df['Average Salary ($)'])
plt.title('Average Salary ($) Over Time')

# Plotting Median Salary ($) over time
plt.subplot(222)
plt.plot(df.index, df['Median Salary ($)'])
plt.title('Median Salary ($) Over Time')

# Plotting Female Average Annual Salary over time
plt.subplot(223)
plt.plot(df.index, df['Female Average Annual Salary'])
plt.title('Female Average Annual Salary Over Time')

# Plotting Male Average Annual Salary over time
plt.subplot(224)
plt.plot(df.index, df['Male Average Annual Salary'])
plt.title('Male Average Annual Salary Over Time')

# Ensuring proper spacing between subplots
plt.tight_layout()

# Showing the plots
plt.show()
```
